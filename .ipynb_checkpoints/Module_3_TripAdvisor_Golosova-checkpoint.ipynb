```python
def capitals(row):
    """Change to binary"""
    if row == 'primary':
        return 1
    if row == 'admin':
        return 0
    
    return row
```

```python
df1['data_1'] = pd.to_datetime(df1['data_1'])
df1['weekend_data_1'] = df1['data_1'].dt.weekday > 4
df1['data_2'] = pd.to_datetime(df1['data_2'])
df1['weekend_data_2'] = df1['data_2'].dt.weekday > 4
df1
```

```python
# Ð Ð°Ð·Ð±Ð¸Ð²Ð°ÐµÐ¼ Ð´Ð°Ñ‚Ð°Ñ„Ñ€ÐµÐ¹Ð¼ Ð½Ð° Ñ‡Ð°ÑÑ‚Ð¸, Ð½ÐµÐ¾Ð±Ñ…Ð¾Ð´Ð¸Ð¼Ñ‹Ðµ Ð´Ð»Ñ Ð¾Ð±ÑƒÑ‡ÐµÐ½Ð¸Ñ Ð¸ Ñ‚ÐµÑÑ‚Ð¸Ñ€Ð¾Ð²Ð°Ð½Ð¸Ñ Ð¼Ð¾Ð´ÐµÐ»Ð¸
# Ð¥ - Ð´Ð°Ð½Ð½Ñ‹Ðµ Ñ Ð¸Ð½Ñ„Ð¾Ñ€Ð¼Ð°Ñ†Ð¸ÐµÐ¹ Ð¾ Ñ€ÐµÑÑ‚Ð¾Ñ€Ð°Ð½Ð°Ñ…, Ñƒ - Ñ†ÐµÐ»ÐµÐ²Ð°Ñ Ð¿ÐµÑ€ÐµÐ¼ÐµÐ½Ð½Ð°Ñ (Ñ€ÐµÐ¹Ñ‚Ð¸Ð½Ð³Ð¸ Ñ€ÐµÑÑ‚Ð¾Ñ€Ð°Ð½Ð¾Ð²)
X = df1.drop([
    'Restaurant_id', 'Rating', 'City', 'Cuisine Style', 'Reviews', 'URL_TA',
    'ID_TA', 'date_of_Review', 'data_2', 'data_1', 'now', 'more_fewer',
    'Vegetarian',  'weekend_data_1', 'weekend_data_2', 
],
             axis=1)
y = df1['Rating']

# Ð—Ð°Ð³Ñ€ÑƒÐ¶Ð°ÐµÐ¼ ÑÐ¿ÐµÑ†Ð¸Ð°Ð»ÑŒÐ½Ñ‹Ð¹ Ð¸Ð½ÑÑ‚Ñ€ÑƒÐ¼ÐµÐ½Ñ‚ Ð´Ð»Ñ Ñ€Ð°Ð·Ð±Ð¸Ð²ÐºÐ¸:
from sklearn.model_selection import train_test_split

# ÐÐ°Ð±Ð¾Ñ€Ñ‹ Ð´Ð°Ð½Ð½Ñ‹Ñ… Ñ Ð¼ÐµÑ‚ÐºÐ¾Ð¹ "train" Ð±ÑƒÐ´ÑƒÑ‚ Ð¸ÑÐ¿Ð¾Ð»ÑŒÐ·Ð¾Ð²Ð°Ñ‚ÑŒÑÑ Ð´Ð»Ñ Ð¾Ð±ÑƒÑ‡ÐµÐ½Ð¸Ñ Ð¼Ð¾Ð´ÐµÐ»Ð¸, "test" - Ð´Ð»Ñ Ñ‚ÐµÑÑ‚Ð¸Ñ€Ð¾Ð²Ð°Ð½Ð¸Ñ.
# Ð”Ð»Ñ Ñ‚ÐµÑÑ‚Ð¸Ñ€Ð¾Ð²Ð°Ð½Ð¸Ñ Ð¼Ñ‹ Ð±ÑƒÐ´ÐµÐ¼ Ð¸ÑÐ¿Ð¾Ð»ÑŒÐ·Ð¾Ð²Ð°Ñ‚ÑŒ 25% Ð¾Ñ‚ Ð¸ÑÑ…Ð¾Ð´Ð½Ð¾Ð³Ð¾ Ð´Ð°Ñ‚Ð°ÑÐµÑ‚Ð°.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
# Ð˜Ð¼Ð¿Ð¾Ñ€Ñ‚Ð¸Ñ€ÑƒÐµÐ¼ Ð½ÐµÐ¾Ð±Ñ…Ð¾Ð´Ð¸Ð¼Ñ‹Ðµ Ð±Ð¸Ð±Ð»Ð¸Ð¾Ñ‚ÐµÐºÐ¸:  
from sklearn.ensemble import RandomForestRegressor # Ð¸Ð½ÑÑ‚Ñ€ÑƒÐ¼ÐµÐ½Ñ‚ Ð´Ð»Ñ ÑÐ¾Ð·Ð´Ð°Ð½Ð¸Ñ Ð¸ Ð¾Ð±ÑƒÑ‡ÐµÐ½Ð¸Ñ Ð¼Ð¾Ð´ÐµÐ»Ð¸  
from sklearn import metrics # Ð¸Ð½ÑÑ‚Ñ€ÑƒÐ¼ÐµÐ½Ñ‚Ñ‹ Ð´Ð»Ñ Ð¾Ñ†ÐµÐ½ÐºÐ¸ Ñ‚Ð¾Ñ‡Ð½Ð¾ÑÑ‚Ð¸ Ð¼Ð¾Ð´ÐµÐ»Ð¸  
  
# Ð¡Ð¾Ð·Ð´Ð°Ñ‘Ð¼ Ð¼Ð¾Ð´ÐµÐ»ÑŒ  
regr = RandomForestRegressor(n_estimators=100)  
  
# ÐžÐ±ÑƒÑ‡Ð°ÐµÐ¼ Ð¼Ð¾Ð´ÐµÐ»ÑŒ Ð½Ð° Ñ‚ÐµÑÑ‚Ð¾Ð²Ð¾Ð¼ Ð½Ð°Ð±Ð¾Ñ€Ðµ Ð´Ð°Ð½Ð½Ñ‹Ñ…  
regr.fit(X_train, y_train)  
  
# Ð˜ÑÐ¿Ð¾Ð»ÑŒÐ·ÑƒÐµÐ¼ Ð¾Ð±ÑƒÑ‡ÐµÐ½Ð½ÑƒÑŽ Ð¼Ð¾Ð´ÐµÐ»ÑŒ Ð´Ð»Ñ Ð¿Ñ€ÐµÐ´ÑÐºÐ°Ð·Ð°Ð½Ð¸Ñ Ñ€ÐµÐ¹Ñ‚Ð¸Ð½Ð³Ð° Ñ€ÐµÑÑ‚Ð¾Ñ€Ð°Ð½Ð¾Ð² Ð² Ñ‚ÐµÑÑ‚Ð¾Ð²Ð¾Ð¹ Ð²Ñ‹Ð±Ð¾Ñ€ÐºÐµ.  
# ÐŸÑ€ÐµÐ´ÑÐºÐ°Ð·Ð°Ð½Ð½Ñ‹Ðµ Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ñ Ð·Ð°Ð¿Ð¸ÑÑ‹Ð²Ð°ÐµÐ¼ Ð² Ð¿ÐµÑ€ÐµÐ¼ÐµÐ½Ð½ÑƒÑŽ y_pred  
y_pred = regr.predict(X_test)  
```

```python
# Ð¡Ñ€Ð°Ð²Ð½Ð¸Ð²Ð°ÐµÐ¼ Ð¿Ñ€ÐµÐ´ÑÐºÐ°Ð·Ð°Ð½Ð½Ñ‹Ðµ Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ñ (y_pred) Ñ Ñ€ÐµÐ°Ð»ÑŒÐ½Ñ‹Ð¼Ð¸ (y_test), Ð¸ ÑÐ¼Ð¾Ñ‚Ñ€Ð¸Ð¼ Ð½Ð°ÑÐºÐ¾Ð»ÑŒÐºÐ¾ Ð¾Ð½Ð¸ Ð² ÑÑ€ÐµÐ´Ð½ÐµÐ¼ Ð¾Ñ‚Ð»Ð¸Ñ‡Ð°ÑŽÑ‚ÑÑ  
# ÐœÐµÑ‚Ñ€Ð¸ÐºÐ° Ð½Ð°Ð·Ñ‹Ð²Ð°ÐµÑ‚ÑÑ Mean Absolute Error (MAE) Ð¸ Ð¿Ð¾ÐºÐ°Ð·Ñ‹Ð²Ð°ÐµÑ‚ ÑÑ€ÐµÐ´Ð½ÐµÐµ Ð¾Ñ‚ÐºÐ»Ð¾Ð½ÐµÐ½Ð¸Ðµ Ð¿Ñ€ÐµÐ´ÑÐºÐ°Ð·Ð°Ð½Ð½Ñ‹Ñ… Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ð¹ Ð¾Ñ‚ Ñ„Ð°ÐºÑ‚Ð¸Ñ‡ÐµÑÐºÐ¸Ñ….  
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
```
